```python
import os
import glob
import pandas as pd

curfolder = os.getcwd()
rawdata = curfolder + '\\rawdata\\'
dataset = curfolder + '\\dataset\\'
```

```python
flesh = 'F:\data_flesh'
responsedata = glob.glob(flesh + '\*\*.csv', recursive=True)

# copy all files to rawdata
for file in responsedata:
    filename = file.split('\\')[-1]
    os.system('copy ' + file + ' ' + rawdata + filename)
    print('copy ' + file + ' ' + rawdata + filename)
```

```python
responsedata = glob.glob(rawdata + '*.csv')

# for now, ignore anything with 18_, 44_, 60_ - NOTE: there has been error during recording so there are two files, we need to merge or exclude
# plus 0_ is pilot dyad 0
responsedata = [file for file in responsedata if '0_' not in file and '18_' not in file and '44_' not in file and '60_' not in file]
```

```python
df_all = pd.DataFrame()

for file in responsedata:
    print('working on ' + file) 

    df = pd.read_csv(file)
    # keep only columns practice, word, modality, answer
    df = df[['practice', 'cycle', 'word', 'modality', 'answer']]
    # get the filename
    filename = file.split('\\')[-1].split('.')[0]
    # first part is session, third part is part of experiment
    sessionID = filename.split('_')[0] + '_' + filename.split('_')[2] 
    sessionID = sessionID.replace('part', '')
    df['exp'] = sessionID.split('_')[1] 
    df['sessionID'] = sessionID

    df_all = pd.concat([df_all, df])

df_all.to_csv(dataset + '\\all_data_raw.csv', index=False)
df_all.head(15)
    
```
